```python
# Import libraries
import pandas as pd
```

```python
# Load the dirty data
df = pd.read_csv("International sale Report.csv")
df.head(10)
```

```python
df.tail(10)
```

```python
# Remove the external index
df.drop(columns = "index", axis = 1, inplace = True)
```

```python
# Check the irregularity in the Date column
df["DATE"].value_counts()[:20]
```

```python
# Some company names are entering the date column
# Duplicate the column and isolate the company names

def remove_date(text):
    # check if the text is not a string
    if not isinstance(text, str):
        text = ""
    # check if there is a hyphen in the text 
    elif "-" in text:
        text = ""
    return text

df["Store Name"] = df["DATE"].apply(remove_date)
df["Store Name"].value_counts()
```

```python
# Some names still have numbers in then and do not appear to be store names

def find_digit (text):
    # remove all the store names with numbers in them
    if any(char.isdigit() for char in text):
        text = ""
    return text

df["Store Name"] = df["Store Name"].apply(find_digit)
```

```python
# Check the column for patterns causing error
df[df["Store Name"] != ""].head(30)
```

*The issue is that this data was joined with another (from row 19675). Upon joining the new data columns did not align properly with the old data columns. An additional data on SKU and Style was also added to the data sheet.*

```python
# Explore the SKU and Style rows
df[18635:19675]
```

*The SKU and Styles have a seperate guide for them and their stocks, remove them and store as a seperate dataset for more analysis.*

```python
# Extract the SKU and Style
SKU_Style = df[18635:19675][["DATE","Months"]]
```

```python
SKU_Style.rename(columns = {
    "DATE": "SKU",
    "MOnths": "Stock"
}, inplace = True)
```

```python
# Remove them from the dataframe and reset index
df.drop(df.index[18635:19675], inplace = True)
df = df.reset_index(drop = True)
```

```python
# Drop Store Name column 
df.drop(columns = "Store Name", axis = 1, inplace = True)
```

```python
#Extract all the data from row 0 to 18635 (before the new rows were added)
original_df = df[:18635]
original_df.tail()
```

```python
# The new rows that were added
new_df = df[18635:]
new_df
```

```python
# Create a new dataframe with all the data from this new df and merge it with the one from the original df
corrected_df = pd.DataFrame()

corrected_df["DATE"] = new_df["Months"]
corrected_df["Months"] = new_df["CUSTOMER"]
corrected_df["CUSTOMER"] = new_df["DATE"]
corrected_df["Style"] = new_df["Style"]
corrected_df["SKU"] = new_df["SKU"]
corrected_df["Size"] = new_df["GROSS AMT"]
corrected_df["PCS"] = new_df["Size"]
corrected_df["RATE"] = new_df["PCS"]
corrected_df["GROSS AMT"] = new_df["RATE"]

corrected_df.head()
```

```python
corrected_df.drop_duplicates().dropna().shape[0]
```

```python
original_df.head()
```

```python
original_df.drop_duplicates().dropna().shape[0]
```

*Both datasets in the new and old columns have similar data, all the columns contain the same data except for the sizes and stock columns in the new corrected data. The data will be merged into one with the stock column aligned with its corresponding value in th original data*

*However, the original data isn't the same size as the ones in the corrected data.*

```python
#Investigate what the corrected data has that the original data doesn't have
merged_df = corrected_df[["SKU","Size"]].merge(
    original_df,
    on = ["SKU"],
    how='outer',
    indicator=True
)

# Check for the values in corrected df and not in the original df
merged_df[merged_df['_merge'] == 'left_only'].isna().sum()
```

```python
# Check for the values only in original df and not in the corrected df
merged_df[merged_df['_merge'] == 'right_only'].isna().sum()
```

*It is best I remove these objects found only in the corrected df*

```python
# The full data without the data solely in the left only
no_left = merged_df[~(merged_df['_merge'] == 'left_only')]
no_left.isna().sum()
```

*These SKU need to be filled with data as removing them will lead to a significant loss in the data*

```python
no_left["SKU"] = no_left["SKU"].fillna(no_left["Style"] + "-" + no_left["Size_y"]) 

# Remove duplicate values and sort by date, this is the useful merged data
merged_df = no_left.drop_duplicates().sort_values("DATE").reset_index(drop = True)

# Remove the _merged column
merged_df.drop(columns = "_merge", inplace = True)
```

```python
# Explore the null values of the merged data
merged_df.isna().sum()
```

```python
# Rename columnns
merged_df.rename(columns = 
{
    "Size_x":"Stock", 
    "Size_y": "Size", 
    "DATE":"Date", 
    "CUSTOMER":"Customer", 
    "RATE":"Rate", 
    "GROSS AMT":"Gross Amt"
}, 
                 inplace = True)
```
